# 1. Sin ve Cos Ã§izimi

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
plt.figure(figsize=(12,6))


t = np.arange(0.0, 2.0, 0.00001)
c = np.cos(-3.14 * np.pi * t)
plt.plot(t,c, linestyle = 'dashed')

t = np.arange(0.0, 2.0, 0.00001)
s = np.sin(1.62 * np.pi * t)
plt.plot(t,-s, linestyle = 'dashed')

plt.xlabel('Zaman (s)')
plt.ylabel('asin+bcos')
plt.title('asin+bcos - Zaman GrafiÄi')

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.savefig('grafik.png')
plt.show()
```

# 3. NotlarÄ±n CSV formatÄ±na getirilmesi ve ilgili grafikler


```python
pd.read_excel("grades.xlsx")
```

```python
finalgrades=[]
for x in range(0,12):
    i = (3*df.iloc[x,2]+3*df.iloc[x,3]+4*df.iloc[x,4])/10
    finalgrades.append(i)
print(finalgrades)
```

```python
df['AÄÄ±rlÄ±klÄ± DÃ¶nem Sonu NotlarÄ±'] = finalgrades
print(df.info())
```

```python
df
```

```python
#u, ortalama deÄeri
u=df['AÄÄ±rlÄ±klÄ± DÃ¶nem Sonu NotlarÄ±'].mean()
print(u)
#s, standart sapma
s=np.std(df['AÄÄ±rlÄ±klÄ± DÃ¶nem Sonu NotlarÄ±'])
print(s)
```

```python
## Verilerimizin normal daÄÄ±ldÄ±ÄÄ±nÄ± ve gÃ¼venilirliÄinin %95 olduÄunu dÃ¼ÅÃ¼nelim. Veri sayÄ±mÄ±z 20'den az olduÄu iÃ§in t testi yapalÄ±m.
## t testimizin formÃ¼lÃ¼ Åu Åekildedir: u+- t.s/n**1/2
## t deÄeri %95 gÃ¼venilirlik deÄerinde ve 12 verimiz olduÄu iÃ§in 11 (12-1) serbestik derecesine sahipken 1.796 deÄeri alÄ±yor.
upperlimit= u + 1.796*s/12**1/2
lowerlimit= u - 1.796*s/12**1/2
print("NotlarÄ±n alt limiti: {}".format(lowerlimit))
```

```python
result=list
df["Passed/You shalll not PASS!"] = result

for x in df.index:
  if df.loc[x, 'AÄÄ±rlÄ±klÄ± DÃ¶nem Sonu NotlarÄ±'] < lowerlimit:
    df.loc[x, "Passed/You shalll not PASS!"] = "You shalll not PASS!"
  else:
    df.loc[x, "Passed/You shalll not PASS!"] = "Passed"
    
```
